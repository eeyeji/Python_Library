```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
from matplotlib import font_manager, rc  
font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
data= pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
#요일별 사고 건수를 세봅시다
temp=data['요일'].value_counts()
temp
```

```python
#노가다필요- 가나다순이기때문에 월~일로는 표현이 안됨
y = temp[['월', '화', '수', '목', '금', '토', '일']]
y
```

```python
x=y.index
plt.bar(x, y, label='사고건수')
plt.legend()

plt.xlabel('요일')
plt.ylabel('사고 건수').set_rotation(0)

#범위 바꿔주기
plt.ylim([500, 600])

plt.title('2017년 요일별 사망교통사고 건수')

plt.show
```

```python
#'사고유형_대분류' 칼럼에서 '차대차'만 뽑아오기
car=data[data['사고유형_대분류']=='차대차']
car
```

```python
#car에서 발생지시도를 기준으로 사상자수의 합계 구하기
#                                    발생지시도 순으로 보겠다.
place=car[['사상자수', '발생지시도']].groupby("발생지시도").sum()
place
```

```python
x=place.index
y=place.values

plt.plot(x,y, label='지역별 사상자수 합계')
plt.legend()

plt.xlabel('지역')
plt.ylabel('사상자수').set_rotation(0)

plt.title('2017년 차vs차 교통사고 사상자수 합계')

plt.show
```
