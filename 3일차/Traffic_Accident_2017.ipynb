```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
from matplotlib import font_manager, rc  
font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
data= pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
#요일별 사고 건수를 세봅시다
temp=data['요일'].value_counts()
temp
```

```python
#노가다필요- 가나다순이기때문에 월~일로는 표현이 안됨
y = temp[['월', '화', '수', '목', '금', '토', '일']]
y
```

```python
x=y.index
plt.bar(x, y, label='사고건수')
plt.legend()

plt.xlabel('요일')
plt.ylabel('사고 건수').set_rotation(0)

#범위 바꿔주기
plt.ylim([500, 600])

plt.title('2017년 요일별 사망교통사고 건수')

plt.show
```

```python
#'사고유형_대분류' 칼럼에서 '차대차'만 뽑아오기
car=data[data['사고유형_대분류']=='차대차']
car
```

```python
#car에서 발생지시도를 기준으로 사상자수의 합계 구하기
#                                    발생지시도 순으로 보겠다.
place=car[['사상자수', '발생지시도']].groupby("발생지시도").sum()
place
```

```python
x=place.index
y=place.values

plt.plot(x,y, label='지역별 사상자수 합계')
plt.legend()

plt.xlabel('지역')
plt.ylabel('사상자수').set_rotation(0)

plt.title('2017년 차vs차 교통사고 사상자수 합계')

plt.show
```

##### 9.23 복습
- 파이차트
- 교통사고가 많이 일어나는 시간대를 알아보고 시각화 하기

```python
#발생년월일시에서 시만 출력하기
time = car.발생년월일시
time%=100
time
```

- 만들어 놓은 time 데이터를 활용하여 카테고리화 시켜주기 - 구간 나누기

```python
#bins: 구간 - 왼쪽초과 오른쪽이하
bins=[-1, 2, 5, 8, 11, 14, 17, 20, 23]

#labels: 구분
labels=["0-2", "3-5", "6-8", "9-11", "12-14", "15-17", "18-20", "21-23"]

result = pd.cut(time, bins, labels=labels)
result
```

- 파이차트 만들기

```python
#시간대에 해당하는 사고건수 세주기
rs = result.value_counts()
rs
```

```python
#                          정렬해주기
rs = result.value_counts().sort_index()
rs
```

```python
plt.figure(figsize=(5,5))

#     데이터   라벨이름                소수점 한자리까지 실수로 표현
#plt.pie(rs, labels=labels, autopct="%1.1f%%")

#                                              시계반대방향
plt.pie(rs, labels=labels, autopct="%1.1f%%", counterclock=False)

plt.show()
```
